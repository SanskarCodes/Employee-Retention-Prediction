```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(
    drop='first',
    sparse_output=False,
    handle_unknown='ignore'
)
```

```python
encoded = encoder.fit_transform(X[["Department", "EducationField", "JobRole","MaritalStatus", "BusinessTravel"]])
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(),
    index=X.index
)
```

```python
X = X.drop(
    columns=[
        "Department",
        "EducationField",
        "JobRole",
        "MaritalStatus",
        "BusinessTravel"
    ]
)
X = pd.concat([X, encoded_df], axis=1)
```

```python
X["Attrition"] = y
```

```python
X.to_csv(
    "/Users/sanskargupta/Documents/Programming /Projects/Employee Retention Prediction Platform/dataset/hr_attrition_encoded.csv",
    index=False
)
```
