```python
import pickle
import pandas as pd

# Load pickle files
with open("/Users/sanskargupta/Documents/Programming /Projects/Employee Retention Prediction Platform/model/best_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("/Users/sanskargupta/Documents/Programming /Projects/Employee Retention Prediction Platform/model/encoder.pkl", "rb") as f:
    encoder = pickle.load(f)

with open("/Users/sanskargupta/Documents/Programming /Projects/Employee Retention Prediction Platform/model/scaler.pkl", "rb") as f:
    scaler = pickle.load(f)

# Sample employee
sample = {
    "Age": 35,
    "DailyRate": 800,
    "DistanceFromHome": 5,
    "Education": 3,
    "EnvironmentSatisfaction": 4,
    "Gender": 1,
    "HourlyRate": 60,
    "JobInvolvement": 3,
    "JobLevel": 2,
    "JobSatisfaction": 4,
    "MonthlyIncome": 7000,
    "MonthlyRate": 15000,
    "NumCompaniesWorked": 2,
    "OverTime": 1,
    "PercentSalaryHike": 15,
    "PerformanceRating": 3,
    "RelationshipSatisfaction": 3,
    "StockOptionLevel": 1,
    "TotalWorkingYears": 10,
    "TrainingTimesLastYear": 2,
    "WorkLifeBalance": 3,
    "YearsAtCompany": 5,
    "YearsInCurrentRole": 3,
    "YearsSinceLastPromotion": 1,
    "YearsWithCurrManager": 3,
    "Department": "Sales",
    "EducationField": "Marketing",
    "JobRole": "Sales Executive",
    "MaritalStatus": "Single",
    "BusinessTravel": "Travel_Rarely"
}

input_df = pd.DataFrame([sample])
```

```python
categorical_cols = [
    "Department",
    "EducationField",
    "JobRole",
    "MaritalStatus",
    "BusinessTravel"
]

encoded = encoder.transform(input_df[categorical_cols])

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(),
    index=input_df.index
)
```

```python
input_df = input_df.drop(columns=categorical_cols)

input_df = pd.concat([input_df, encoded_df], axis=1)
```

```python
input_df = input_df.reindex(columns=model.feature_names_in_, fill_value=0)
```

```python
input_scaled = scaler.transform(input_df)
```
