```python
scaler = pickle.load(open("/Users/sanskargupta/Documents/Programming /Projects/Employee Retention Prediction Platform/model/scaler.pkl", "rb"))

print(len(scaler.feature_names_in_))
```

```python
with open("/Users/sanskargupta/Documents/Programming /Projects/Employee Retention Prediction Platform/model/encoder.pkl", "rb") as f:

    encoder = pickle.load(f)
```

```python
import pandas as pd
sample = pd.DataFrame({

    "Age": [35],

    "DailyRate": [800],

    "DistanceFromHome": [5],

    "Education": [3],

    "EnvironmentSatisfaction": [4],

    "Gender": [1],

    "HourlyRate": [60],

    "JobInvolvement": [3],

    "JobLevel": [2],

    "JobSatisfaction": [4],

    "MonthlyIncome": [7000],

    "MonthlyRate": [15000],

    "NumCompaniesWorked": [2],

    "OverTime": [1],

    "PercentSalaryHike": [15],

    "PerformanceRating": [3],

    "RelationshipSatisfaction": [3],

    "StockOptionLevel": [1],

    "TotalWorkingYears": [10],

    "TrainingTimesLastYear": [2],

    "WorkLifeBalance": [3],

    "YearsAtCompany": [5],

    "YearsInCurrentRole": [3],

    "YearsSinceLastPromotion": [1],

    "YearsWithCurrManager": [3],

    "Department": ["Sales"],

    "EducationField": ["Marketing"],

    "JobRole": ["Sales Executive"],

    "MaritalStatus": ["Single"],

    "BusinessTravel": ["Travel_Rarely"]

})
```

```python
cat_cols = [

    "Department",

    "EducationField",

    "JobRole",

    "MaritalStatus",

    "BusinessTravel"

]
```

```python
encoded = encoder.transform(sample[cat_cols])

encoded_df = pd.DataFrame(

    encoded,

    columns=encoder.get_feature_names_out(),

    index=sample.index

)
```

```python
sample = sample.drop(columns=cat_cols)
```

```python
sample = pd.concat([sample, encoded_df], axis=1)
```

```python
num_cols = list(scaler.feature_names_in_)
```

```python
sample[num_cols] = scaler.transform(sample[num_cols])
```

```python
sample = sample.reindex(columns=model.feature_names_in_, fill_value=0)
```

```python
prediction = model.predict(sample)

probability = model.predict_proba(sample)

print("Prediction :", prediction[0])

if prediction[0] == 1:

    print("Employee is likely to leave.")

else:

    print("Employee is likely to stay.")

print("\nProbability of Staying :", round(probability[0][0] * 100, 2), "%")

print("Probability of Leaving :", round(probability[0][1] * 100, 2), "%")
```
